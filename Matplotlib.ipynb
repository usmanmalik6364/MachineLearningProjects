<h1>(Matplotlib example workflow)</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
#Using object oriented 

fig,ax = plt.subplots(figsize = (10,10))

#3. Plot data
ax.plot(x,y)

#4 Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

#5 Save & show

fig.savefig('Matplots/sampleplot.png')
```

## Making figures with Numpy Arrays

```python
#Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
#Plot the data and create a line plot. The default is line plot
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#Use same data to make a scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#Make a plot from dictionary
nut_butter_prices = {'Almond Butter': 10,
                    'Peanut Butter': 8,
                    'Cashew Butter': 12}
fig,ax = plt.subplots()
ax.set(title='Nut butter Store',
      ylabel='Price ($)')
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
fig,ax = plt.subplots()
nut_butter_prices = {'Almond Butter': 10,
                    'Peanut Butter': 8,
                    'Cashew Butter': 12}
ax.set(title='Nut butter Store',
      xlabel='Price ($)')
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
#Histogram
#make some data
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

<h2>subplots</h2>

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows =2,
                                        ncols = 2,
                                        figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(10000));
```
