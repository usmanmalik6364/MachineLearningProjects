<h1>(Matplotlib example workflow)</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
#Using object oriented 

fig,ax = plt.subplots(figsize = (10,10))

#3. Plot data
ax.plot(x,y)

#4 Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

#5 Save & show

fig.savefig('sampleplot.png')
```
